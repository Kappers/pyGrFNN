```python
%matplotlib inline

from numpy.polynomial.polynomial import polyroots

alpha, beta1, beta2, epsilon = -5, 10, -3, 1  # double limit-cycle
# alpha, beta1, beta2, epsilon = 0, -5, -1, 1  # limit-cycle

F = np.arange(0, 2, 0.2)
# F = [0]
r = np.arange(0, 1/np.sqrt(epsilon), 0.01)
rdot = np.add.outer(alpha * r + beta1 * r**3 + ((epsilon* beta2 * r**5)/(1 - epsilon * r**2)), F)


# colors = [colormap(i) for i in np.linspace(0, 0.7, len(F))]
colors = [jhColors(i) for i in np.linspace(0, 1.0, len(F))]

plt.clf()
ax = plt.gca()
ax.set_color_cycle(colors)
plt.plot(r, rdot, zorder=0)
# assymptote
# plt.vlines(x=1/np.sqrt(epsilon), ymin=-1, ymax=2, color='r', linestyle=':')
plt.ylim(-2,2)
ax.axhline(y=0,xmin=min(r),xmax=max(r),c="k",zorder=5, alpha=0.5)
# plt.grid(True)
plt.title(r'$\alpha={:.3g}, \quad \beta_1={:.3g}, \quad \beta_2={:.3g}, \quad \varepsilon={:.3g}$'.format(alpha, beta1, beta2, epsilon))

# find roots
roots = [None] * len(F)
for i in xrange(len(F)):
#     r = np.roots([epsilon*(beta2-beta1), 0, beta1-epsilon*alpha, -epsilon*F[i], alpha, F[i]])
    r = polyroots([F[i], alpha, -epsilon*F[i], beta1-epsilon*alpha, 0, epsilon*(beta2-beta1)])
    r = np.real(r[np.abs(np.imag(r)) < 1e-20])
    r = r[(r>=0) & (r < 1/np.sqrt(epsilon))]
    roots[i] = r
# print roots

plt.gca().set_color_cycle(colors)
for r in roots:
    plt.plot(r, np.zeros_like(r), 'o', markersize=4, zorder=10)
    
    
# tikz_save('/Users/jorgeh/Documents/CCRMA/thesis/tikzFigures/double-limit-cycle.tex', figureheight='5cm', figurewidth='8cm')
# tikz_save('/Users/jorgeh/Documents/CCRMA/thesis/tikzFigures/limit-cycle.tex', figureheight='5cm', figurewidth='8cm')
```
